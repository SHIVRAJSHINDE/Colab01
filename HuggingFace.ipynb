```python
from transformers import pipeline

```

```python
senti_model = pipeline(task="sentiment-analysis")
```

```python
senti_model("this is a good movie, but screenplay is not good")
```

```python
newModel = "cardiffnlp/twitter-roberta-base-sentiment-latest"
senti_model01 = pipeline(task="sentiment-analysis",model=newModel,device='cuda')
senti_model01
```

```python
senti_model01("this is a good movie")
```

```python
senti_model01("Overheating issue, dont buy product")
```

```python
label = senti_model01("Overheating issue, dont buy product")[0]['label']
label
```

```python
# Apply it on the data

import pandas as pd
user_review_data=pd.read_csv("https://raw.githubusercontent.com/venkatareddykonasani/Datasets/master/Amazon_Yelp_Reviews/Review_Data.csv")
user_review_data=user_review_data.sample(50)
user_review_data["Review"]
```

```python
def get_label(x):
    return senti_model01(x)[0]["label"]

user_review_data['LABEL'] = user_review_data['Review'].apply(get_label)
user_review_data
```

```python
# Question Answering

qa_Model = pipeline(task="question-answering",model="deepset/roberta-base-squad2",device="cuda")

qa_Model
```
